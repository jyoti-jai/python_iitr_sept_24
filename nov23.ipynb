```python
'''This is a constructor'''
class Vector:
    # it does something
    '''This is a constructor'''
    def __init__(self, x, y):
        if type(x) != int:
            print("ERROR!")
            return None
        self.x = x
        self.y = y
    def norm(self):
        x, y = self.x, self.y
        return math.sqrt(x*x + y * y)
    def add(self, other):
        xx = self.x + other.x
        yy = self.y + other.y
        return Vector(xx, yy)
    def __repr__(self):
        return f"-->[{self.x}, {self.y}]"
```

```python
help(Vector)
```

```python
v1 = Vector("a", "b")
```

```python
v1 = Vector(3, 4)
```

```python
v1.norm() # norm(v)
```

```python
v2 = Vector(5, 6)
```

```python
v3 = v1.add(v2)
```
