```python
# Write your own add digit array. Part a)
#arr1=['2','3','4']
#arr2=['8','7','5']

def addNumbers(arr1, arr2):
    i = len(arr1) - 1
    j=len(arr2) - 1
    
    carryOver = 0
    result =[]
    
    while (i>=0 or j>=0):
        if (i>=0):
            num1 = myInt(arr1[i])
        else:
            num1 = 0
        
        if (j>=0):
            num2 = myInt(arr2[j])
        else:
            num2 = 0
        i-=1
        j-=1
        numAdd = num1+num2+carryOver
        result.append(numAdd %10)
        carryOver = numAdd // 10
    
    if carryOver > 0 :
        result.append(carryOver)
    result.reverse()
    return result
```

```python
# calcmethod= 0 for max, 1 for min, any other value for avg
def scaledown(img, calcmethod=0):
    newimg=[]
    n=2
    height= len(img)
    width = len(img[0])
    newh = int(height/n)
    neww = int(width/n)
    newimg = [[0]*neww for _ in range(newh)]
    for r in range(newh):
        for c in range(neww):
            pixelvalues=[]
            for i in range(n):
                j=c*n
                k=(r*n)+i
                pixelvalues+= img[k][j:j+n]
                # totalval+= sum(img[k][j:j+n])
            if (calcmethod==0):
                newval = max(pixelvalues)
            elif (calcmethod ==1):
                newval = min(pixelvalues)
            else:
                # newval = totalval/(n*n) # taking the average of the cells
                newval = sum(pixelvalues)/(n*n)
                
            newimg[r][c]=newval
    return newimg   
```
